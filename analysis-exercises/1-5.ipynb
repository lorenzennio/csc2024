```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
def CDF(pdf, lo=-np.inf, hi=np.inf):
    def cdf(x):
        return sp.integrate.quad(pdf, lo, x)[0]
    return cdf

pdf = lambda x: 1/2 if 0 <= x <= 1 else 0
cdf = CDF(pdf)

x = np.linspace(-1, 2, 100)

plt.plot(x, [pdf(xi) for xi in x])

plt.plot(x, [cdf(xi) for xi in x])
```

```python
pdf = sp.stats.norm.pdf

cdf = CDF(pdf)

x = np.linspace(-5, 5, 100)

plt.plot(x, [pdf(xi) for xi in x])
plt.plot(x, [cdf(xi) for xi in x])
```

```python
def PDF(cdf, dx=1e-9):
    def pdf(x):
        return (cdf(x + dx) - cdf(x - dx))/(2*dx)
    return pdf

cdf = sp.stats.norm.cdf

pdf = PDF(cdf)

x = np.linspace(-5, 5, 100)

plt.plot(x, [cdf(xi) for xi in x])
plt.plot(x, [pdf(xi) for xi in x])
```
